```python
import inv_competition as ic
import pandas as pd
import numpy as np
import datetime

today = datetime.date.today()
yesterday = today - datetime.timedelta(days=1)
last_Friday = today - datetime.timedelta(days=3) # 월요일에는 yesterday 대신 last_Friday를 input
```

```python
before = ic.import_sheet(date=yesterday)
```

```python
# 마지막 column의 형태를 yyyy-mm-dd로 바꿔줌
# inv_competition.py 파일의 import_sheet method에 해당 코드를 추가하였으나, 위에 before.head(3)을 보면 먹히지 않았음 (수정 필요)
cols = list(before.columns)
cols[-1] = yesterday
before.columns = cols
before.head(3)
```

```python
code_list = ic.get_code_list(before)
print('today code list:', code_list)
print('개수:', str(len(code_list))+'개')
```

```python
scrap = ic.get_price(code_list) # scraping the current prices
scrap_df = pd.DataFrame(scrap)
update = before.merge(scrap_df, how='outer') # merge before and current data by code
update.head(3)
```

```python
ic.fill(update) # filling prices(NaN) if there are some holders
ic.cal_rates(update) # calculate profit rates
final = ic.ordering(update, '수익률')
final.head(3)
```

```python
# 평균수익률 계산
total_rate = round(np.mean(final['수익률']), 2)
print("전체포트수익률: " + str(total_rate) + '%')
```

```python
ic.export_sheet(final)
```
